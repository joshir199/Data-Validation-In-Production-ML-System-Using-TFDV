```python
# Describing data with TFDV
# The usual workflow when using TFDV during training is as follows:
"""
1. Generate statistics for the data
2. Use those statistics to generate a schema for each feature
3. Visualize the schema and statistics and manually inspect them
4. Update the schema if needed
"""

"""
TFDV can compute descriptive statistics that provide a quick overview of the data
in terms of the features that are present and the shapes of their value distributions.

Internally, TFDV uses Apache Beam's data-parallel processing framework to scale
the computation of statistics over large datasets.
"""

# NOTE: Compute statistics

# tfdv.generate_statistics_from_csv
# tfdv.generate_statistics_from_dataframe
# tfdv.generate_statistics_from_tfrecord
```

```python
# Compute data statistics for the input training data from csv.
stats_from_csv = tfdv.generate_statistics_from_csv('sample_data/california_housing_train.csv')

# Visualize statistics
tfdv.visualize_statistics(stats_from_csv)
```

```python
# Compute data statistics for the input training data from dataframe.
stats_from_dataframe = tfdv.generate_statistics_from_dataframe(housing_data_train)

# Visualize statistics
tfdv.visualize_statistics(stats_from_dataframe)
```

```python
# TFDV generates following statistics for the features

"""
TFDV generates different types of statistics based on the type of features.
For numerical features, TFDV computes for every feature:

-> Count of records
-> Number of missing (i.e. null values)
-> Histogram of values
-> Mean and standard deviation
-> Minimum and maximum values
-> Percentage of zero values


For categorical features, TFDV provides:

-> Count of values
-> Percentage of missing values
-> Number of unique values
-> Average string length
-> Count for each label and its rank

"""
```

```python
# Lets compare train and tests data statistics:

train_stats = tfdv.generate_statistics_from_dataframe(dataframe=housing_data_train)
test_stats = tfdv.generate_statistics_from_dataframe(dataframe=housing_data_test)

tfdv.visualize_statistics(
  lhs_statistics=train_stats, lhs_name='TRAIN_DATASET',
  rhs_statistics=test_stats, rhs_name='NEW_DATASET')
```

```python
# Defining the Schema for data:
"""
Now let's use tfdv.infer_schema to create a schema for our data. A schema defines
constraints for the data that are relevant for ML. Example constraints include
the data type of each feature, its possible range, whether it's numerical or
categorical, or the frequency of its presence in the data.

Getting the schema right is important because the rest of our production pipeline
will be relying on the schema that TFDV generates to be correct.
"""
# Generating Schema
# Once statistics are generated, the next step is to generate a schema for our dataset.
# This schema will map each feature in the dataset to a type (float, bytes, etc.).
# Also define feature boundaries (min, max, distribution of values and missings, etc.).

# Infers schema from the input statistics.
schema = tfdv.infer_schema(statistics=train_stats)
print(schema)
```

```python
# The schema also provides documentation for the data, and so is useful when different
# developers work on the same data.
tfdv.display_schema(schema)
```

```python
"""
In this visualization, the columns stand for:

-> Type indicates the feature datatype.

-> Presence indicates whether the feature must be present in 100% of examples (required) or not (optional).

-> Valency indicates the number of values required per training example.

-> Domain and Values indicates The feature domain and its values

Note: In the case of categorical features, valency = single indicates that each
      training example must have exactly one category for the feature.

"""
```

```python
# TFDV lets you update the schema according to your domain knowledge of the data
# if you are not satisfied by the auto-generated schema.

# 1. Change optional features to required.
households_feature = tfdv.get_feature(schema, 'households')
households_feature.presence.min_fraction = 1.0
```

```python
# 2. Update(add/remove) a feature with a new value. mostly useful in Categorical data
# 3. Change a feature from a float to an integer
total_room_feature = tfdv.get_feature(schema, 'total_rooms')
total_room_feature.type = 2
tfdv.display_schema(schema)
```

```python
# Check for evaluation anomalies
# Does our evaluation dataset match the schema from our training dataset?
# This is especially important for categorical features, where we want to identify
# the range of acceptable values or numeric features that are outside the ranges.

test_data_anomalies = tfdv.validate_statistics(statistics=test_stats, schema=schema)
tfdv.display_anomalies(anomalies=test_data_anomalies)
```

```python
# Fix evaluation anomalies in the schema
total_room_feature = tfdv.get_feature(schema, 'total_rooms')
total_room_feature.type = 3  # for FLOAT

updated_anomalies = tfdv.validate_statistics(test_stats, schema)
tfdv.display_anomalies(updated_anomalies)
```

```python
# Schema Environments
# By default all datasets in a pipeline should use the same schema, but there are often exceptions.
"""
Environments can be used to express such requirements. In particular, features in
schema can be associated with a set of environments using default_environment,
in_environment and not_in_environment.
"""

# For example, In Serving dataset, we do not need label column as we will be predicting
# label using model. So, this column will be missing from its schema
serving_stats = tfdv.generate_statistics_from_dataframe(housing_data_serving)
serving_anamolies = tfdv.validate_statistics(serving_stats, schema)
tfdv.display_anomalies(serving_anamolies)

```

```python
# In this scenario, we can set the schema environment variable to ignore this column
# All features are by default in both TRAINING and SERVING environments.
schema.default_environment.append('TRAINING')
schema.default_environment.append('SERVING')

# Specify that 'median_house_value' feature is not in SERVING environment.
tfdv.get_feature(schema, 'median_house_value').not_in_environment.append('SERVING')

# check again the anamolies
serving_anamolies_with_env = tfdv.validate_statistics(serving_stats, schema, environment='SERVING')
tfdv.display_anomalies(serving_anamolies_with_env)
```

```python
# Check for drift and skew in the SERVING dataset
# TFDV performs this check by comparing the statistics of the different datasets
# based on the drift/skew comparators specified in the schema.

"""
Data drift is expressed in terms of Chebyshev distance (also L-infinity distance)
which calculates distance between two vectors in vector space.
We can set the Thresold distance and measure if it exceeds the thresold or not.
"""
```

```python
# Skew Measurement
# TFDV can detect three kinds of skew in the data - schema skew, feature skew, distribution skew.

"""
1. Schema :
     Both training and serving data are expected to adhere to the same schema.
     Any expected deviations between the two is considered as data skew.

2. Feature :
     Feature skew occurs when the feature values that a model trains on are different
     from the feature values that it sees at serving time.

3. Distribution :
     Distribution skew occurs when the distribution of the training dataset is significantly
     different from the distribution of the serving dataset. It can be because of different
     code used or different sampling used for training and serving data
"""
```

```python
# Lets calculates drift and skew in the serving data
median_income_feature = tfdv.get_feature(schema, "median_income")
median_income_feature.skew_comparator.infinity_norm.threshold = 0.00001

housing_median_age_type = tfdv.get_feature(schema, "housing_median_age")
housing_median_age_type.drift_comparator.infinity_norm.threshold = 0.00001

skew_drift_anomalies = tfdv.validate_statistics(train_stats, schema,
                                          previous_statistics=test_stats,
                                          serving_statistics=serving_stats)
```

```python
# display anamolies
tfdv.display_anomalies(skew_drift_anomalies)
```

```python
# Freeze the schema
# Once the schema is reviewed and all the anamolies are resolved or put
# within thresold, We can save(or freeze) the schema as a frozen state.

from tensorflow.python.lib.io import file_io
from google.protobuf import text_format
import os

OUTPUT_DIR = "sample_data/schema_output_file"
file_io.recursive_create_dir(OUTPUT_DIR)
schema_file = os.path.join(OUTPUT_DIR, 'schema.pbtxt')
tfdv.write_schema_text(schema, schema_file)
```
